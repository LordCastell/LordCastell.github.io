```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("violencia.xlsx", usecols=["Grupo de edad de la victima", "País de Nacimiento de la Víctima"])

df_colombia = df[df["País de Nacimiento de la Víctima"] == "Colombia"]
top_edades = df_colombia["Grupo de edad de la victima"].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_edades.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 10 de Rangos de Edad con más Casos de Violencia de Género en Colombia")
plt.xlabel("Grupo de edad de la victima")
plt.ylabel("Cantidad de Casos")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Top 10 de Rangos de Edad con más Casos de Violencia de Género en Colombia**

La gráfica muestra los 10 grupos de edad más afectados por la violencia de género en Colombia. Se observa que el rango "25 a 29 años" presenta la mayor cantidad de casos, seguido por "20 a 24 años" y "30 a 34 años", lo que indica que las personas jóvenes, particularmente entre los 20 y 34 años, son las más vulnerables a este tipo de violencia.

Después de estos grupos, los casos disminuyen progresivamente en los siguientes rangos de edad. Sin embargo, aún se reportan cantidades significativas en personas de 35 a 39 años, 40 a 44 años y 45 a 49 años.

Es importante resaltar que también hay registros de violencia en edades más tempranas, como los grupos de 10 a 14 años y 15 a 17 años, lo que evidencia que este problema afecta incluso a menores de edad.

Estos datos resaltan la necesidad de políticas de prevención y protección dirigidas especialmente a la población joven y a menores en situación de vulnerabilidad.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("violencia.xlsx", usecols=["Año del hecho", "Sexo de la victima"])
casos_por_año = df.groupby(["Año del hecho", "Sexo de la victima"]).size().unstack()
plt.figure(figsize=(12, 6))
casos_por_año.plot(kind='bar', stacked=False, color=['pink', 'lightblue'], edgecolor='black')


plt.title("Casos de Violencia de Género por Año (Hombres vs. Mujeres)")
plt.xlabel("Año")
plt.ylabel("Cantidad de Casos")
plt.xticks(rotation=45)
plt.legend(title="Sexo de la víctima")
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

### **Análisis de la Gráfica: Casos de Violencia de Género por Año (Hombres vs. Mujeres)**  

La gráfica muestra la cantidad de **casos de violencia de género** en los últimos años, diferenciando entre **hombres y mujeres** como víctimas.  

📌 **Tendencias Generales:**  
- Se observa que, en todos los años analizados, la mayoría de las víctimas son **mujeres** (barras en azul claro), representando una proporción significativamente mayor en comparación con los hombres.  
- Los casos han mantenido una tendencia relativamente estable en algunos años, aunque con fluctuaciones notables.  
- En **2019 y 2020**, hay una caída en el número de casos, posiblemente relacionada con restricciones de movilidad o cambios en la denuncia de estos hechos.  
- A partir de **2021**, se observa un aumento progresivo en la cantidad de denuncias, con un pico en **2023**, año con el mayor número de casos registrados.  

📊 **Diferencia por Género:**  
- Aunque los hombres también son víctimas de violencia de género, la cantidad de casos en su contra es mucho menor en comparación con las mujeres.  
- La proporción de casos en **hombres** se mantiene relativamente constante a lo largo de los años, sin cambios drásticos.  

📢 **Conclusión:**  
El análisis muestra que la violencia de género sigue siendo un problema grave, especialmente para las mujeres, quienes representan la mayoría de las víctimas. La tendencia de aumento en los últimos años resalta la importancia de **políticas de prevención, apoyo a las víctimas y concientización social** para reducir estos casos en el futuro.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("violencia.xlsx", usecols=["Mecanismo Causal", "Sexo de la victima"])
casos_por_año = df.groupby(["Mecanismo Causal", "Sexo de la victima"]).size().unstack()

plt.figure(figsize=(12, 6))
casos_por_año.plot(kind='bar', stacked=False, color=['pink', 'lightblue'], edgecolor='black') # Colors in the correct order

# Customize the plot
plt.ylabel("Cantidad de Casos")
plt.title("Casos de Violencia de Género - Mecanismo Causal (Hombres vs. Mujeres)")
plt.xlabel("Mecanismo Causal")
plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotate and adjust font size
plt.legend(title="Sexo de la víctima")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent overlapping

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("violencia.xlsx", usecols=["Presunto Agresor"])
# Calculate the size of each group and reset the index
casos_por_año = df.groupby(["Presunto Agresor"]).size().reset_index(name='counts')

#Now you have a DataFrame with 'Presunto Agresor' and 'counts' columns

plt.figure(figsize=(12, 6))
plt.bar(casos_por_año['Presunto Agresor'], casos_por_año['counts'], color=['skyblue'], edgecolor='black') # Use plt.bar directly

# Customize the plot
plt.ylabel("Cantidad de Casos")
plt.title("Casos de Violencia de Género - Presunto Agresor")
plt.xlabel("Presunto Agresor")
plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotate and adjust font size

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent overlapping

plt.show()
```
